```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_2015.drop(['Sale (Dollars)'],axis=1), 
                                                    data_2015['Sale (Dollars)'], test_size=0.33, 
                                                    random_state=42)
```

```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score


# Don't forget to turn on plotting display in the notebook
%matplotlib inline 

lm = linear_model.LinearRegression()

X = X_train[['Store Number','Bottles Sold']]
y = y_train

#y -= np.mean(y)
#y /= np.std(y)

# model = lm.fit(X, y)

# predictions  =  model.predict(X)
# score        =  model.score(X, y)

# print(model.coef_)
# print('\n\nScore: {0}'.format(score))


scores = cross_val_score(lm,X,y)
print('avg r2:',scores.mean())
print('std r2:',scores.std())
```

```python
lm.fit(X, y)
```

```python
predictions  =  lm.predict(X_test[['Store Number','Bottles Sold']])
score        =  lm.score(X_test[['Store Number','Bottles Sold']], y_test)

# Plot the model
plt.figure(figsize=(8,8))
plt.scatter(predictions, y_test, s=30, c='r', marker='+', zorder=10)
plt.xlabel("Predicted Values")
plt.ylabel("Actual Values")

plt.plot([0, np.max(y)], [0, np.max(y)])

plt.show()

print("score: ", score)
```
